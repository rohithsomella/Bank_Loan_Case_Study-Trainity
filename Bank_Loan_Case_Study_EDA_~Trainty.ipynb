Here we should also consider the percentage level of missing values in each column If there is a huge number of missing values in a single column and if we wanted to remove with respect to rows of entire data set it definitely affects negatively on the data set because of one single column contains huge number of missing values

```python
missing_values_count = X.isnull().sum() # storing the count of missing values in each column

total_rows = X.shape[0] # total number of rows

percentage_missing_values = (missing_values_count / total_rows) * 100 # no. of missing values /total no. of rows

# dataframe for the clear undestanding
missing_values_df = pd.DataFrame({
    'Column Name': missing_values_count.index,
    'Missing Values Count': missing_values_count.values,
    'Percentage Missing': percentage_missing_values.values
})

missing_values_df = missing_values_df.sort_values(by='Percentage Missing', ascending=False) # high percentage

missing_values_df.reset_index(drop=True, inplace=True) #replacing the index and droping the original column

pd.set_option('display.max_columns', None) # visibility of 120 columns
pd.set_option('display.max_rows', None)

df = pd.DataFrame(missing_values_df) # rearrenging the values into 1 dataframe
df
```

Removing Columns having >50% of the null values

```python

missing_values_count = X.isnull().sum() # finding missing values of each column
total_rows = X.shape[0] # total number of rows
percentage_missing_values = (missing_values_count / total_rows) * 100 # caluating the percentage of missing's


columns_to_remove = percentage_missing_values[percentage_missing_values > 50].index # threshold 50% to remove

X = X.drop(columns=columns_to_remove) # removing the column contains >50% of missing values

```

```python
X.info() # column names and their data types
```

Now we have only 50 colums in the dataset after removing the higher missing values containing columns

Now we can remove the missing values with respect to rows

```python
col = list(X.columns) # now all columns in a list to handle null values
col
```

```python
X.isnull().sum()
```

filling null values which contains less than 50% of null values with centeral tendency

```python
for i in X.columns:
    if X[i].dtypes == 'object':
        X[i] = X[i].fillna(X[i].mode()[0])
    else:
        X[i] = X[i].fillna(X[i].median())
```
